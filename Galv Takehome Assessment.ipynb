```python
#Challenge 4

def count_isograms(wordlist):
    isolist = []
    for word in wordlist:
        isolist.append(isisogram(word))
    return isolist.count(True)

"""
The main part of this function just takes in a list of words and then passes
each word into the isisogram helper function below. When the result of that
function is returned for every item in wordlist, the count of the amount of
times that True is handed back from isisogram in that list is returned as the
result for count_isograms.
"""

def isisogram(word):
    boollist = []
    caseword = word.lower()
    
    for letter in caseword:
        
        if caseword.count(letter) == 1:
            boollist.append(True)
            
        else:
            boollist.append(False)
    
    if False in boollist:
        return False
    
    else:
        return True
    
"""
The isisogram helper function made it a little bit easier for me to stay organized
while I was writing count_isograms above. It takes in a single word (so one position
from the list that was passed in to count_isogram.) and checks to see if it is an
isogram. First, I set the word to be all lowercase so that I am not leaving myself
open to any possible errors due to case. Then, for each letter in caseword, I check
the count of that letter is 1, if so, I append a True value to a list of booleans
representing if each letter is only in a word once. Otherwise, I would append false.
The last thing isisogram does is that it checks to see if any position in boollist
has the value False, and if so, it returns False to count_isograms. Otherwise (so if
a word IS an isogram) it will return True to the main function.
"""
```
